```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from tqdm import tqdm
import time
import pickle
import requests
import os
import pathlib
```

```python
driver = webdriver.Chrome('../../chromedriver')
```

```python
WEBSITE_PAUSE_TIME = 1
IMAGE_PAUSE_TIME = 0.5
SCROLL_PAUSE_TIME = 2
```

```python
image_urls=[]
viewed_pages = set()

queue = [
    'https://www.uhdpaper.com/2020/04/sunrise-nature-lake-flowers-landscape.html'
]
```

```python
def scroll_to_bottom(driver):
    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        
        
```

```python
i = 0
while True:
    if len(queue) <= 0:
        break
        
    url = queue.pop(0)
    
    if url in viewed_pages:
        continue
    
    driver.get(url)
    
    
    img_link = driver.find_element(By.XPATH, '//div[@class="wp_container"]//picture/img')
    img_url = img_link.get_attribute('src')
    
    break
    
    img_slug = img_url.split('/')[-1]
    path = pathlib.Path('/data1', 'Raw', 'UHDWallpapers', img_slug)
    path.parent.mkdir(parents=True, exist_ok=True)
    with open(path,'wb') as f:
        f.write(requests.get(img_url).content)
        f.close()
    
    
    prev_link = driver.find_elements(By.XPATH, '//a[@id="Blog1_blog-pager-newer-link"]')
    next_link = driver.find_elements(By.XPATH, '//a[@id="Blog1_blog-pager-older-link"]')
    
    if len(prev_link) > 0:
        prev_url = prev_link[0].get_attribute('href')
        queue.append(prev_url)
        
    if len(next_link) > 0:
        next_url = next_link[0].get_attribute('href')
        queue.append(next_url)
    
    viewed_pages.add(url)
    
    i+=1
    if (i%50) == 0:
        print(i)
        with open('/data1/uhdPhotos.pkl', 'wb') as f:
            pickle.dump(image_urls, f)
            
    time.sleep(WEBSITE_PAUSE_TIME)
```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import pyautogui as pg
from tqdm import tqdm
import time
import pickle
import requests
import os
import pathlib
import time
```

```python
WEBSITE_PAUSE_TIME = .1
IMAGE_PAUSE_TIME = .1
SCROLL_PAUSE_TIME = .1
```

```python
image_urls=[]
viewed_pages = set()

queue = [
    'https://www.uhdpaper.com/2020/04/sunrise-nature-lake-flowers-landscape.html'
]
```

```python
def scroll_to_bottom(driver):
    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        
        
```

```python
downloaded_before = False
def save_img():
    global downloaded_before
    
    y = 1742-950
    if downloaded_before:
        y-=45
    pg.click(970+32,y, button='middle')
    time.sleep(IMAGE_PAUSE_TIME)
    pg.click(1300, 75)
    time.sleep(IMAGE_PAUSE_TIME)
    pg.keyDown('ctrl')
    pg.press('s')
    pg.keyUp('ctrl')
    time.sleep(IMAGE_PAUSE_TIME)
    pg.click(1500, 950)
    time.sleep(2*IMAGE_PAUSE_TIME)
    pg.keyDown('ctrl')
    pg.press('w')
    pg.keyUp('ctrl')
    time.sleep(IMAGE_PAUSE_TIME)
    downloaded_before=True
```

```python
driver = webdriver.Chrome('../../chromedriver')
```

```python
limit = 10000
i = 0
while True:
#     t1=time.time()
    if len(queue) <= 0:
        break
        
    if i > limit:
        break
        
    url = queue.pop(0)
    
    if url in viewed_pages:
        continue
    
    driver.get(url)
    
    scroll_to_bottom(driver)
    WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, '//div[@class="wp_container"]//picture/img')))
    save_img()
    
    prev_link = driver.find_elements(By.XPATH, '//a[@id="Blog1_blog-pager-newer-link"]')
    next_link = driver.find_elements(By.XPATH, '//a[@id="Blog1_blog-pager-older-link"]')
    
    if len(prev_link) > 0:
        prev_url = prev_link[0].get_attribute('href')
        queue.append(prev_url)
        
    if len(next_link) > 0:
        next_url = next_link[0].get_attribute('href')
        queue.append(next_url)
    
    viewed_pages.add(url)
    
    i+=1
    if (i%50) == 0:
        print(i)
        with open('/data1/uhdPhotos.pkl', 'wb') as f:
            pickle.dump(image_urls, f)
            
    
#     print(time.time() -t1)
```
